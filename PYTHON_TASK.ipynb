====================

Student Performance Analysis Using Python Arrays
-----------------------------------------------

PROMPT: Create a Python program using arrays for a Student Performance Analysis system.

Requirements:
- Create an array of student marks
- Display all marks
- Access specific values
- Update values
- Find highest mark
- Find lowest mark
- Calculate average


PROMPT: Install array module

```python
from array import array
```

PROMPT: Create Student Marks Array

```python
marks = array('i',[85,90,72,88,95])

print("Student Marks:")
print(marks)
```

PROMPT: Access Array Elements

```python
print("First Student Mark:", marks[0])

print("Third Student Mark:", marks[2])
```

PROMPT: Update Array Value

```python
marks[2] = 78

print("Updated Marks:")
print(marks)
```

PROMPT: Find Highest and Lowest Marks

```python
highest = max(marks)

lowest = min(marks)

print("Highest Mark:", highest)

print("Lowest Mark:", lowest)
```

PROMPT: Calculate Average

```python
average = sum(marks)/len(marks)

print("Average Mark:", average)
```

PROMPT: Sort Array

```python
marks = array('i',sorted(marks))

print("Sorted Marks:")
print(marks)
```

PROMPT: Search Student Mark

```python
search = 90

if search in marks:
    print(search,"is present")
else:
    print(search,"not found")
```

PROMPT: Filter Above Average Students

```python
average = sum(marks)/len(marks)

print("Students above average:")

for mark in marks:
    if mark > average:
        print(mark)
```
